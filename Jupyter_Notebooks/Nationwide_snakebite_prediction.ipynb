```python
import numpy as np
import pandas as pd

ap_data = pd.read_csv("../data/APdata.csv")
all_states = pd.read_csv("../data/All_states.csv")
```

```python
ap_data.head()
```

```python
all_states.head()
```

```python
all_states['total_no_of_bites'] = 0
for a in range(0,len(all_states)):
    if all_states['States'][a] == 'Andhra Pradesh':
        all_states.loc[a,'total_no_of_bites'] = sum(i for i in ap_data['no_of_bites'])

all_states
```

```python
all_states['total_no_of_deaths'] = 0
for a in range(0,len(all_states)):
    if all_states['States'][a] == 'Andhra Pradesh':
        all_states.loc[a,'total_no_of_deaths'] = sum(i for i in ap_data['no_of_deaths'])

all_states
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

corr = all_states.corr()
sns.heatmap(corr)
```

```python
Obs = all_states.loc[all_states['States'] == 'Andhra Pradesh']
```

```python
X_train = Obs[['rural_pop','total_area','human_density_psqkm','sex_ratio_p1000m','literacy','forest_cover_sqkm']]
y_train = Obs[['total_no_of_bites','total_no_of_deaths']]
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
print('Coefficients: \n', lm.coef_)
```

```python
Samp_1 = all_states.loc[all_states['States'] == 'Tamil Nadu']
X_test = Samp_1[['rural_pop','total_area','human_density_psqkm','sex_ratio_p1000m','literacy','forest_cover_sqkm']]
y_test = Samp_1[['total_no_of_bites','total_no_of_deaths']]
```

```python
y_pred = lm.predict(X_test)
```
